## Scrapping de quelques films de IMDB

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
start_time = time.time()
from selenium import webdriver


def imdb_scrap_ex():
    """Return a dataframe with informations from the titles scrapped from IMDB"""
    #lists of values that we will put in a dataframe
    ratings=[]
    rating_counts=[]
    titles=[]
    audiences=[]
    creators=[]
    nb_comments=[]
    countries=[]
    release_dates=[]
    filming_locations=[]
    sound_mixes=[]
    colors=[]
    aspect_ratios=[]
    budgets=[]
    opening_weekend=[]
    gross_us=[]
    worldwide_gross=[]
    genres=[]
    #IMDB url of the titles
    
    f_url="https://www.imdb.com/search/title/?num_votes=5000,&sort=user_rating,desc&start=301"



    while(True):
      #getting html code from the url
        try:
            r_imdb=requests.get(f_url)
        except:
            print("request exception" )
            dico={"ratings":ratings,"rating_counts":rating_counts,"titles":titles,"audiences":audiences,"creators":creators,"nb_comments":nb_comments,"countries":countries,"release_dates":release_dates,"filming_locations":filming_locations,"sound_mixes":sound_mixes,"colors":colors,"aspect_ratios":aspect_ratios,"budgets":budgets,"opening_weekend":opening_weekend,"worldwide_gross":worldwide_gross,"gross_us":gross_us,"genres":genres}

            return pd.DataFrame(dico)
        soup = BeautifulSoup(r_imdb.text, 'html.parser')
        #results contains the reference to every title on the page
        results = soup.find_all('div', attrs={'class':'lister-item mode-advanced'})
        for result in results:
          #getting the url of a title
            href=result.find("a")["href"]
            url="https://www.imdb.com"+href
            

            


           
            try:
              driver.get(url)
            except:
                print("request exception")
                continue

            #html parsing
            html = driver.page_source
            soup_film = BeautifulSoup(html, 'html.parser')
            #getting ratings
            print("ratings")
            
            rating="XXX"
            tmp = soup_film.find('span', attrs={'itemprop':'ratingValue'})
            if(tmp !=None):
                rating=tmp.text
                
    

            #getting rating counts
            rating_count="XXX"
            tmp = soup_film.find('span', attrs={'itemprop':'ratingCount'})
            if(tmp !=None):
                rating_count=tmp.text
                
            #getting titles
            title="XXX"
            tmp = soup_film.find('h1')
            if(tmp !=None):
                title=tmp.text
                
            

            #getting audience
            audience="XXX"
            tmp = soup_film.find('div',attrs={'class':'subtext'})
            try:
                tmp=audience.text.split('\n')[1].strip()
                if(tmp !=None and tmp!=''):
                    audience=tmp
                    
            except:
                print("exception: no audience")
                
            #getting creator
            creator="XXX"
            tmp=soup_film.find('div',attrs={'class':'credit_summary_item'})
            try:
                tmp=creator.find('a').text
                if(tmp !=None):
                    creator=tmp
                    
            except:
                print("exception: no creator")
                
            #getting the number of comments
            nb_comm="XXX"
            tmp=soup_film.find('div',attrs={'class':'user-comments'})
            try:
                tmp=tmp.find_all('a',attrs={'onclick':''})
                tmp=tmp[2].text.split()[2]
                if(tmp !=None):
                    nb_comm=tmp
            except:
                print("exception: cant get comments number")
                
            #getting countries
            details=soup_film.find('div',attrs={'id':'titleDetails'})
            details_found=False
            try:
                tmp=details.find_all()
                details_found=True
            except:
              print("details not found")

            country="XXX"
            date="XXX"
            location="XXX"
            mix="XXX"
            color="XXX"
            ratio="XXX"
            budget="XXX"
            opening="XXX"
            gross="XXX"
            cumulative="XXX"


            if(details_found):
                try:
                    for i,v in enumerate(tmp):
                        if("Country" in v.text):
                            tmp2=tmp[i].text.split("\n")[2]
                            if(tmp2!=None):
                                country=tmp2
                            

                                break
                
                except:
                    print("country not found")            
            #getting release dates

                try:
                    for i,v in enumerate(tmp):
                        if("Release Date" in v.text):
                            string=" "
                            tmp3=tmp[i].text
                            if(tmp3!=None):
                                date=string.join(tmp3.split()[2:5])
                                print(date)

                                break
                except:
                    print("release date not found") 

            #getting filming locations
                try:
                    for i,v in enumerate(tmp):
                        if("Filming" in v.text):
                            tmp4=tmp[i].find('a').text
                            if(tmp4!=None):
                                location=tmp4
                            
                                break
                except:
                    print("filming location not found") 
            
            #getting sound_mixes
                try:
                    for i,v in enumerate(tmp):
                        if("Sound Mix" in v.text):
                            tmp5=tmp[i].find('a').text
                            if(tmp5!=None):
                                mix=tmp5
                            
                                break
                
                except:
                    print("filming location not found") 

            #getting colors
                try:
                    for i,v in enumerate(tmp):
                        if("Color" in v.text):
                            tmp6=tmp[i].find('a').text
                            if(tmp6!=None):
                                color=tmp6
                            
                                break
                
                except:
                    print("color not found")
            #getting aspect ratios
                try:
                    for i,v in enumerate(tmp):
                        if("Aspect Ratio" in v.text):
                            ratio=tmp[i].text.replace("Aspect Ratio:","")                            
                            break
                    
                except:
                    print("aspect ratio not found")
            #get budget
                try:
                    for i,v in enumerate(tmp):
                        if("Budget" in v.text):
                            budget=tmp[i].text.replace("Budget:","").replace("estimated","").replace("()","").replace("$","")
                            break

                    
                except:
                    print("budget not found")
                #Opening USA
                try:
                    for i,v in enumerate(tmp):
                        if("Opening Weekend" in v.text):
                            opening=tmp[i].text.replace("Opening Weekend USA:","").replace("estimated","").replace("()","").replace("$","")
                        break
                except:
                    print("opening not found")

                try:
                    for i,v in enumerate(tmp):
                        if("Gross USA" in v.text):
                            gross=tmp[i].text.replace("Gross USA:","").replace("estimated","").replace("()","").replace("$","").replace(" ","")
                            break
                    
                except:
                    print("gross us not found")

                try:
                    for i,v in enumerate(tmp):
                        if("Cumulative" in v.text):
                            cumulative=tmp[i].text.replace("CumulativeWorldwideGross:","").replace("estimated","").replace("()","").replace("$","").replace(" ","")
                            break
                    
                except:
                    print("gross us not found")
            
            #get genres
            genre="XXX"
            storyLine=soup_film.find('div',attrs={'id':'titleStoryLine'})
            storyLine_found=False
            try:
                tmp=storyLine.find_all()
                storyLine_found=True
            except:
                print("storyline not found")
            
            if(storyLine_found):
                
                try:
                    for i,v in enumerate(tmp):
                        if("Genres" in v.text):
                            genre=tmp[i].text.replace("Genres:","").replace("|" ,"").strip()
                            print(genre)
                            break
                    
                except:
                    print("genre not found")
        
            ratings.append(rating)
            rating_counts.append(rating_count)
            titles.append(title)
            audiences.append(audience)
            creators.append(creator)
            nb_comments.append(nb_comm)
            countries.append(country)
            release_dates.append(date)
            filming_locations.append(location)  
            sound_mixes.append(mix)
            colors.append(color)
            aspect_ratios.append(ratio)
            budgets.append(budget)
            opening_weekend.append(opening)
            gross_us.append(gross)
            worldwide_gross.append(cumulative)
            genres.append(genre)
            
      #getting the url of the next page so we can start scrapping the next titles    
    
        next_page = soup.find('a', attrs={'class':'lister-page-next next-page'})
        try:
           f_url="https://www.imdb.com"+next_page["href"]
           print(f_url)
           
        except:
            dico={"ratings":ratings,"rating_counts":rating_counts,"titles":titles,"audiences":audiences,"creators":creators,"nb_comments":nb_comments,"countries":countries,"release_dates":release_dates,"filming_locations":filming_locations,"sound_mixes":sound_mixes,"colors":colors,"aspect_ratios":aspect_ratios,"budgets":budgets,"opening_weekend":opening_weekend,"worldwide_gross":worldwide_gross,"gross_us":gross_us,"genres":genres}

            return pd.DataFrame(dico)   
            
    dico={"ratings":ratings,"rating_counts":rating_counts,"titles":titles,"audiences":audiences,"creators":creators,"nb_comments":nb_comments,"countries":countries,"release_dates":release_dates,"filming_locations":filming_locations,"sound_mixes":sound_mixes,"colors":colors,"aspect_ratios":aspect_ratios,"budgets":budgets,"opening_weekend":opening_weekend,"worldwide_gross":worldwide_gross,"gross_us":gross_us,"genres":genres}
      
    return pd.DataFrame(dico)

df_ex=imdb_scrap_ex()
print("--- %s seconds ---" % (time.time() - start_time))
df_ex.to_csv("imdbwithgenresfirst.csv")
```

## Pre Traitement

```python
df_1=pd.read_csv("imdbwithgenresfirst.csv")
df_2=pd.read_csv("imdbwithgenres3.csv")
df_3=pd.read_csv("imdbwithgenres4.csv")
```

```python
df=pd.concat([df_2,df_3])
```

```python
df.head()
```

```python
#nettoyage colonne budget
df['budgets'] = df['budgets'].apply(lambda x: x.replace("GBP",""))
```

```python
#nettoyage colonne budget
df['budgets'] = df['budgets'].apply(lambda x: x.strip())
```

```python
import numpy as np
def to_int(x):
    try:
        return int(x.replace(',',''))
    except:
        return np.nan
```

```python
#transformer la colonne budget en int
df['budgets'] = df['budgets'].apply(to_int)
```

```python
#nettoyage colonne genres
df['genres'] = df['genres'].apply(lambda x: x.strip())
```

```python
df['gross_us'] = df['gross_us'].apply(lambda x: x.strip())
```

```python
def worldwide_strip(x):
  try:
    return x.replace("CumulativeWorldwideGross:","").strip()
  except:
    return x
```

```python
#nettoyage colonnes revenus  generes
df['worldwide_gross']=df['worldwide_gross'].apply(worldwide_strip)
```

```python
df['aspect_ratios']=df['aspect_ratios'].apply(lambda x: x.strip())
```

```python
#conversion colonne revenus en int
df['worldwide_gross']=df['worldwide_gross'].apply(to_int)
```

```python
df['gross_us']=df['gross_us'].apply(to_int)
```

```python
def to_float(x):
    try:
        return float(x.replace(',','.'))
    except:
        return np.nan
```

```python
#conversion des notes en float
df["ratings"]=df["ratings"].apply(to_float)
```
